```python
# Store results for cross-country comparison
all_cleaned_dfs = []
country_stats = {}

for country_name, file_name in COUNTRIES:
    print(f"\n{'='*70}")
    print(f"PROCESSING: {country_name}")
    print(f"{'='*70}")
    
    # --- LOAD DATA ---
    path = os.path.join('..', file_name)
    
    if not os.path.exists(path):
        print(f"File not found: {path}. Skipping {country_name}.")
        continue
    
    df = pd.read_csv(path)
    df['Country'] = country_name
    df['Date'] = pd.to_datetime(df['YEAR'] * 1000 + df['DOY'], format='%Y%j')
    df['Month'] = df['Date'].dt.month
    df['Year'] = df['Date'].dt.year
    
    print(f"Data loaded: {len(df)} rows")
    print(f"Date range: {df['Date'].min().date()} to {df['Date'].max().date()}")
    
    # --- REPLACE SENTINEL VALUES ---
    df.replace(-999, np.nan, inplace=True)
    print("Replaced -999 sentinel values with NaN")
    
    # --- DUPLICATE CHECK ---
    dup_count = df.duplicated().sum()
    print(f"Duplicate rows found: {dup_count}")
    if dup_count > 0:
        df.drop_duplicates(inplace=True)
        print(f"Duplicates removed. New shape: {df.shape}")
    
    # --- MISSING VALUE REPORT ---
    missing_count = df.isna().sum()
    missing_pct = (missing_count / len(df)) * 100
    missing_report = pd.DataFrame({
        'Missing Count': missing_count,
        'Percentage': missing_pct.round(2)
    })
    missing_report = missing_report[missing_report['Missing Count'] > 0].sort_values('Percentage', ascending=False)
    
    if len(missing_report) > 0:
        print("Missing Value Report:")
        print(missing_report)
        
        high_missing = missing_report[missing_report['Percentage'] > 5]
        if len(high_missing) > 0:
            print("Columns with >5% missing values:")
            for col in high_missing.index:
                print(f"  - {col}: {high_missing.loc[col, 'Percentage']:.1f}%")
    
    # --- DESCRIPTIVE STATISTICS ---
    print("Descriptive Statistics:")
    print(df[NUMERIC_COLS].describe().round(3))
    
    # --- OUTLIER DETECTION ---
    z_scores = np.abs(stats.zscore(df[OUTLIER_COLS].dropna()))
    outlier_mask = (z_scores > 3).any(axis=1)
    outlier_count = outlier_mask.sum()
    outlier_pct = (outlier_count / len(df)) * 100
    
    print(f"Outlier Detection (|Z| > 3): {outlier_count} rows ({outlier_pct:.1f}%)")
    
    # Cap outliers at 3 standard deviations
    for col in OUTLIER_COLS:
        mean_val = df[col].mean()
        std_val = df[col].std()
        upper_bound = mean_val + 3 * std_val
        lower_bound = mean_val - 3 * std_val
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    
    # --- HANDLE MISSING VALUES ---
    df[OUTLIER_COLS] = df[OUTLIER_COLS].ffill()
    
    # Drop rows with >30% missing values
    row_missing_pct = df.isna().sum(axis=1) / len(df.columns)
    rows_to_drop = (row_missing_pct > 0.3).sum()
    if rows_to_drop > 0:
        df = df[row_missing_pct <= 0.3]
        print(f"Dropped {rows_to_drop} rows with >30% missing values")
    
    print(f"Final cleaned shape: {df.shape}")
    
    # --- SAVE CLEANED DATA ---
    output_file = os.path.join(OUTPUT_DIR, f"{country_name.lower()}_clean.csv")
    df.to_csv(output_file, index=False)
    print(f"Saved: {output_file}")
    
    # Store for later use
    all_cleaned_dfs.append(df)
    country_stats[country_name] = {
        'mean_temp': df['T2M'].mean(),
        'std_temp': df['T2M'].std(),
        'mean_rain': df['PRECTOTCORR'].mean(),
        'std_rain': df['PRECTOTCORR'].std(),
        'extreme_heat_days': (df['T2M_MAX'] > 35).sum(),
        'outlier_pct': outlier_pct
    }

print(f"\n{'='*70}")
print(f"ALL COUNTRIES PROCESSED: {len(all_cleaned_dfs)} datasets saved")
```

```python
# Plot monthly mean temperature for each country
fig, axes = plt.subplots(len(all_cleaned_dfs), 1, figsize=(16, 4 * len(all_cleaned_dfs)))

if len(all_cleaned_dfs) == 1:
    axes = [axes]

for idx, df in enumerate(all_cleaned_dfs):
    country_name = df['Country'].iloc[0]
    
    monthly_df = df.set_index('Date').resample('M')['T2M'].mean()
    
    axes[idx].plot(monthly_df.index, monthly_df.values, color='crimson', marker='.', markersize=4, linewidth=1.5)
    
    warmest_date = monthly_df.idxmax()
    warmest_val = monthly_df.max()
    coolest_date = monthly_df.idxmin()
    coolest_val = monthly_df.min()
    
    axes[idx].annotate(
        f'Warmest: {warmest_val:.1f}C\n{warmest_date.strftime("%b %Y")}',
        xy=(warmest_date, warmest_val),
        xytext=(15, 15), textcoords='offset points',
        fontsize=9, color='darkred',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
        arrowprops=dict(arrowstyle='->', color='darkred')
    )
    
    axes[idx].annotate(
        f'Coolest: {coolest_val:.1f}C\n{coolest_date.strftime("%b %Y")}',
        xy=(coolest_date, coolest_val),
        xytext=(15, -25), textcoords='offset points',
        fontsize=9, color='darkblue',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='lightblue', alpha=0.7),
        arrowprops=dict(arrowstyle='->', color='darkblue')
    )
    
    axes[idx].set_title(f'Monthly Mean Temperature — {country_name}', fontsize=13, fontweight='bold')
    axes[idx].set_ylabel('Temperature (C)')
    axes[idx].grid(True, alpha=0.3)
    
    # Trend line
    x_numeric = np.arange(len(monthly_df))
    z = np.polyfit(x_numeric, monthly_df.values, 1)
    p = np.poly1d(z)
    axes[idx].plot(monthly_df.index, p(x_numeric), '--', color='gray', alpha=0.7, 
                   label=f'Trend: {z[0]*12:.3f}C/year')
    axes[idx].legend(loc='upper left', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(len(all_cleaned_dfs), 1, figsize=(16, 4 * len(all_cleaned_dfs)))

if len(all_cleaned_dfs) == 1:
    axes = [axes]

for idx, df in enumerate(all_cleaned_dfs):
    country_name = df['Country'].iloc[0]
    
    monthly_rain = df.set_index('Date').resample('M')['PRECTOTCORR'].sum()
    
    axes[idx].bar(monthly_rain.index, monthly_rain.values, color='steelblue', width=20, alpha=0.8)
    
    peak_date = monthly_rain.idxmax()
    peak_val = monthly_rain.max()
    
    axes[idx].annotate(
        f'Peak: {peak_val:.0f} mm\n{peak_date.strftime("%b %Y")}',
        xy=(peak_date, peak_val),
        xytext=(15, 15), textcoords='offset points',
        fontsize=9, fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='lightgreen', alpha=0.8),
        arrowprops=dict(arrowstyle='->')
    )
    
    axes[idx].set_title(f'Monthly Total Precipitation — {country_name}', fontsize=13, fontweight='bold')
    axes[idx].set_ylabel('Total Precipitation (mm)')
    axes[idx].grid(True, alpha=0.3, axis='y')
    
    mean_rain = monthly_rain.mean()
    axes[idx].axhline(y=mean_rain, color='red', linestyle='--', alpha=0.7, 
                      label=f'Mean: {mean_rain:.0f} mm/month')
    axes[idx].legend(loc='upper right', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
for df in all_cleaned_dfs:
    country_name = df['Country'].iloc[0]
    corr_matrix = df[NUMERIC_COLS].corr()
    
    corr_pairs = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    corr_pairs = corr_pairs.unstack().dropna()
    
    top_3 = corr_pairs.abs().sort_values(ascending=False).head(3)
    
    print(f"Top 3 Strongest Correlations — {country_name}")
    for (var1, var2), abs_val in top_3.items():
        actual_val = corr_matrix.loc[var1, var2]
        print(f"  {var1} vs {var2}: r = {actual_val:.3f}")
    print()
```

```python
### Correlation Analysis Interpretation

**Recurring Strong Correlations:**

1. **T2M_MAX vs T2M (r ≈ 0.9+)**: Expected relationship — mean and maximum temperatures are highly correlated.

2. **T2M vs RH2M (strongly negative)**: Warmer temperatures drive lower relative humidity. This amplifies drought stress on vegetation.

3. **T2M_RANGE vs RH2M**: Larger daily temperature swings are associated with drier air.
```
